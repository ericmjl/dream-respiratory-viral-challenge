```python
import pandas as pd
```

```python
expression_df = pd.read_csv('../data/ViralChallenge_training_EXPRESSION_RMA.tsv', sep="\t", index_col='FEATUREID').T
expression_df.shape
```

```python
clinical_df = pd.read_csv('../data/ViralChallenge_training_CLINICAL.tsv', sep='\t')
clinical_df.set_index('CEL', inplace=True)
clinical_df = clinical_df[clinical_df['TIMEHOURS'] <= 24]
# clinical_df.set_index(['STUDYID'])
clinical_df
```

```python
# Split the data by study ID
clinical_split = dict()
for grp, dat in clinical_df.groupby('STUDYID'):
    print(grp)
    if grp == 'Rhinovirus Duke':
        print('->passing!')
        pass
    else:
        clinical_split[grp] = dat.join(expression_df).reset_index().sort_values(['STUDYID', 'SUBJECTID', 'SHEDDING_SC1', 
                                                                                 'SYMPTOMATIC_SC2', 'LOGSYMPTSCORE_SC3', 
                                                                                 'AGE', 'GENDER', 'EARLYTX', 'SHAM', 'TIMEHOURS'])
        del clinical_split[grp]['SAMPLEID']
        del clinical_split[grp]['CEL']

        clinical_split[grp] = clinical_split[grp].set_index(['STUDYID', 'SUBJECTID', 'SHEDDING_SC1', 
                                                             'SYMPTOMATIC_SC2', 'LOGSYMPTSCORE_SC3', 
                                                             'AGE', 'GENDER', 'EARLYTX', 'SHAM', 'TIMEHOURS'])
        clinical_split[grp] = clinical_split[grp].unstack()
        clinical_split[grp].columns = clinical_split[grp].columns.to_series().apply(pd.Series).astype(str).T.apply('_'.join)
        clinical_split[grp].to_csv('{0}.csv'.format(grp))
```

```python
clinical_split['DEE2 H3N2']
```

```python
combined_df = clinical_df.join(expression_df).reset_index().sort_values(['STUDYID', 'SUBJECTID', 'SHEDDING_SC1', 'SYMPTOMATIC_SC2', 'LOGSYMPTSCORE_SC3', 'AGE', 'GENDER', 'EARLYTX', 'SHAM', 'TIMEHOURS'])
combined_df = combined_df.set_index()
combined_df = combined_df.unstack()

combined_df.columns = combined_df.columns.to_series().apply(pd.Series).astype(str).T.apply('_'.join)
combined_df.reset_index().drop('level_0')
```

```python
combined_df.to_csv('ViralChallenge_training_COMBINED_TRANSFORMED.csv')
```

```python
# Create a new pandas dataframe from dictionaries.
new_data = []
for (studyid, subjectid), dat in combined_df.groupby(['STUDYID', 'SUBJECTID']):
    datum = dict()
    datum['STUDYID'] = studyid
    datum['SUBJECTID'] = subjectid
    datum['AGE'] = np.unique(dat['AGE'].values)[0]
    datum['GENDER'] = np.unique(dat['GENDER'].values)[0]
    datum['EARLYTX'] = np.unique(dat['EARLYTX'].values)[0]
    datum['SHAM'] = np.unique(dat['SHAM'].values)[0]
    datum['SHEDDING_SC1'] = np.unique(dat['SHEDDING_SC1'].values)[0]
    datum['SYMPTOMATIC_SC2'] = np.unique(dat['SYMPTOMATIC_SC2'].values)[0]
    datum['LOGSYMPTSCORE_SC3'] = np.unique(dat['LOGSYMPTSCORE_SC3'].values)[0]
    
    
    for r, d in dat.iterrows():
        time = str(d['TIMEHOURS'])
        for col in expression_df.columns:
            newcol = str(col) + ' | ' + time
            datum[newcol] = d[col]
    new_data.append(datum)
```

```python
pd.DataFrame(new_data).to_csv('ViralChallenge_training_COMBINED_TRANSFORMED.csv')
```

```python
feat_cols = expression_df.columns
out_cols = ['SHEDDING_SC1']

X = combined_df[feat_cols]
Y = combined_df['SHEDDING_SC1']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split, StratifiedShuffleSplit, cross_val_score

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

sss = StratifiedShuffleSplit(Y, n_iter=10, test_size=0.2)

cv_scores = cross_val_score(RandomForestClassifier(n_estimators=300), X, Y, cv=sss)
```

```python
cv_scores.mean(), cv_scores.std()
```

```python
rfc = RandomForestClassifier(n_estimators=400)
rfc.fit(X_train, Y_train)
rfc.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(Y_test, rfc.predict(X_test))
```
